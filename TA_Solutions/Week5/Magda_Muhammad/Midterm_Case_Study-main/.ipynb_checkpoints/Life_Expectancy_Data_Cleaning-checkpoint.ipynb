```python
def headers_to_lowercase (df):

    df.columns = df.columns.str.lower()
    
    return df
```

```python
def text_to_lowercase (df):
    
    df = df.applymap(lambda s: s.lower() if type(s) == str else s)
    
    return df   
```

```python
life_df = headers_to_lowercase(life_df)
life_df = text_to_lowercase(life_df)
```

```python
def drop_columns (df, list_of_columns_to_drop):
    df.drop(columns = list_of_column_to_drop, inplace=True)
    return df

list_of_columns_to_drop = ['country_code', 'hospitals','gni_capita','une_hiv','une_gni', 'une_poverty', 'une_edu_spend', 'une_literacy',
       'une_school']    
```

```python
def fill_na_in_population_with_items_from_dict (df, dictionary, country_filter):
    df['une_pop'][df['country'] == country_filter] = df['une_pop'][df['country'] == country_filter].fillna(df['year'].map(dictionary), inplace=True)
    return df
```
