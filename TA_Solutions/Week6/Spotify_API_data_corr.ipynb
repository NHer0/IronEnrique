```python
# getting track ids - simmilar to the version we tried

def getTrackIDs_corr(user, playlist_id):

    ids = []
    playlist = sp.user_playlist(user,playlist_id,market="GB")["tracks"]  # we go to a lower level of the dictionary using key=tracks, 
                                                                         # there we have the key=next
                                                                         
    for item in playlist['items']:  # we include a first for loop outside the while to get the avoid skipping the first 100 songs

        track = item['track']
        ids.append(track['id'])


    while playlist["next"]:

        playlist = sp.next(playlist)  # now sp.next find the key=next and is able to page up to the next 100 tracks

        for item in playlist['items']:

            track = item['track']
            ids.append(track['id'])

    return ids

#playlist 1 
ids1 = getTrackIDs_corr('22xt5wnkv63vf2spvsxcpkexa', '6K5xXt4BZyzo9PkYwIwjUO')
```

```python
#get features for each track
def getTrackFeatures(id):
  meta = sp.track(id)
  features = sp.audio_features(id)

  # meta
  name = meta['name']
  album = meta['album']['name']
  artist = meta['album']['artists'][0]['name']
  release_date = meta['album']['release_date']
  length = meta['duration_ms']
  popularity = meta['popularity']

  # features
  acousticness = features[0]['acousticness']
  danceability = features[0]['danceability']
  energy = features[0]['energy']
  instrumentalness = features[0]['instrumentalness']
  liveness = features[0]['liveness']
  loudness = features[0]['loudness']
  speechiness = features[0]['speechiness']
  tempo = features[0]['tempo']
  time_signature = features[0]['time_signature']

  track = [name, album, artist, release_date, length, popularity, danceability, acousticness, energy, instrumentalness, liveness, loudness, speechiness, tempo, time_signature]
  return track
```

```python
# loop over track ids 
import time
tracks = []
for i in range(len(ids1)):
  time.sleep(.5)
  track = getTrackFeatures(ids1[i])
  tracks.append(track)

# create dataset
df1 = pd.DataFrame(tracks, columns = ['name', 'album', 'artist', 'release_date', 'length', 'popularity', 'danceability', 'acousticness',  'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'time_signature'])
```
